#Which channel generates most conversions?
Which channel has highest conversion rate?
Which campaigns perform best?
Channel contribution to conversions

```python
#Channel-wise Conversion Count
channel_conversion = (
    df[df['Conversion'] == 'Yes']
    .groupby('Channel')
    .size()
    .sort_values(ascending=False)
)

print(channel_conversion)
```

```python
#Channel Conversion Rate
channel_summary = df.groupby('Channel').agg(
    Total_Visits=('User ID', 'count'),
    Conversions=('Conversion',
                 lambda x: (x == 'Yes').sum())
)

channel_summary['Conversion_Rate'] = (
    channel_summary['Conversions']
    / channel_summary['Total_Visits']
) * 100

channel_summary
```

```python
#Campaign Analysicampaign_summary = df.groupby('Campaign').agg(
campaign_summary = df.groupby('Campaign').agg(
    Total_Visits=('User ID','count'),
    Conversions=('Conversion',
                 lambda x:(x=='Yes').sum())
)

campaign_summary.sort_values(
    'Conversions',
    ascending=False
)
```

```python
#Visualizations

df['Channel'].value_counts().plot(kind='bar')
plt.title('Traffic by Channel')
plt.show()

```

Data Quality Assessment

The provided Web Analytics dataset was inspected for missing values and marketing attribution fields.

Findings:

* No missing values were detected in the available columns.
* The dataset contains the following fields:

  * User ID
  * Timestamp
  * Channel
  * Campaign
  * Conversion
* UTM parameters (utm_source, utm_medium, utm_campaign, utm_term, utm_content) were not present in the source dataset.

Action Taken:

* Verified dataset completeness using Pandas null-value analysis.
* Confirmed no UTM fields were available for standardization or imputation.
* Documented dataset limitations for future attribution model development.

Conclusion:
No UTM cleaning was required because UTM-related attributes were not provided in the source data.

```python
# Find datetime-related columns

for col in df.columns:
    if 'time' in col.lower() or 'date' in col.lower():
        print(col)
```

```python
#Convert to Datetime Format
df['Timestamp'] = pd.to_datetime(
    df['Timestamp'],
    errors='coerce'
)
```

```python
##Channel Distribution Chart
#Purpose
#Show which marketing channels generate the most traffic.
import matplotlib.pyplot as plt

channel_counts = df['Channel'].value_counts()

plt.figure(figsize=(8,5))
channel_counts.plot(kind='bar')

plt.title('Channel Distribution')
plt.xlabel('Channel')
plt.ylabel('Number of Touchpoints')

plt.tight_layout()

plt.savefig('../docs/channel_distribution.png')

plt.show()
save_path = '../report/images/channel_Distribution.png'
plt.savefig('../report/images/channel_Distribution.png',
            bbox_inches='tight')
print("successful")
```

```python
#Conversion Chart
#Shows converted vs non-converted records.
conversion_counts = df['Conversion'].value_counts()

plt.figure(figsize=(6,4))

conversion_counts.plot(
    kind='bar'
)

plt.title('Conversion Distribution')
plt.xlabel('Conversion')
plt.ylabel('Count')

plt.tight_layout()

plt.savefig('../docs/conversion_distribution.png')

plt.show()
save_path = '../report/images/channel_conversion.png'
plt.savefig('../report/images/channel_conversion.png',
            bbox_inches='tight')
print("successful")
```

```python
#Channel-wise conversions.
channel_conversion = (
    df[df['Conversion']=='Yes']
    .groupby('Channel')
    .size()
)

plt.figure(figsize=(8,5))

channel_conversion.plot(kind='bar')

plt.title('Conversions by Channel')
plt.xlabel('Channel')
plt.ylabel('Conversions')

plt.tight_layout()

plt.savefig('../docs/channel_conversion.png')

plt.show()
save_path = '../report/images/Conversions by Channel.png'
plt.savefig('../report/images/Conversions by Channel.png',
            bbox_inches='tight')
print("successful")
```

```python
#. Traffic Trend Chart
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
daily_traffic = (
    df.groupby(
        df['Timestamp'].dt.date
    )
    .size()
)

plt.figure(figsize=(12,5))

daily_traffic.plot()

plt.title('Daily Traffic Trend')
plt.xlabel('Date')
plt.ylabel('Visits')

plt.tight_layout()

plt.savefig('../docs/traffic_trend.png')

plt.show()
save_path = '../report/images/Daily Traffic Trend.png'
plt.savefig('../report/images/Daily Traffic Trend.png',
            bbox_inches='tight')
print("successful")
```
